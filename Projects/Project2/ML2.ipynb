Distribution plot

```python
def plotting(var,num):
    plt.subplot(2,2,num)
    sns.histplot(df[var],kde = True)

plotting('Age',1)
plotting('RestingBP',2)
plotting('Cholesterol',3)
plotting('MaxHR',4)


plt.tight_layout()
```

Short explanation:
def plotting(var, num): → Creates a reusable function to plot a histogram.
plt.subplot(2,2,num) → Arranges plots in a 2 × 2 grid.
sns.histplot(df[var], kde=True) → Draws a histogram with a KDE (smooth density curve).
plotting('Age',1) → Plots Age distribution.
plotting('RestingBP',2) → Plots Resting Blood Pressure distribution.
plotting('Cholesterol',3) → Plots Cholesterol distribution.
plotting('MaxHR',4) → Plots Maximum Heart Rate distribution.
plt.tight_layout() → Adjusts spacing so the four plots don't overlap

```python
df["Cholesterol"].value_counts()
```

calculate mean cholestroll excluding 0 values (becz 0 is invalid)

```python
ch_mean=df.loc[df['Cholesterol']!=0,'Cholesterol'].mean()
```

replace 0 with mean

```python
df['Cholesterol']=df["Cholesterol"].replace(0,ch_mean)
```

round decimal value to 2

```python
df['Cholesterol']=df["Cholesterol"].round(2)
```

```python
resting_bp_mean = df.loc[df['RestingBP'] != 0, 'RestingBP'].mean()

df['RestingBP'] = df['RestingBP'].replace(0, resting_bp_mean)

df['RestingBP'] = df['RestingBP'].round(2)
```

replot histogram after cleaning

```python
def plotting(var,num):
    plt.subplot(2,2,num)
    sns.histplot(df[var],kde=True)
plotting("Age",1)
plotting("RestingBP",2) 
plotting("Cholesterol",3) 
plotting("MaxHR",4) 
plt.tight_layout()
```
